# Intro to Data Science

## What is data science?
Data science can be many things but mostly it is

* Machine learning
* Statistics
* Experimental design
* Explaining statistics to business
* Using business understanding to phrase reasonable data-answerable questions

Depending on your background and why you wish to learn Data Science you will have different angles to the topic. This course is designed for two different directions.

1. Those who want to learn more about data science with the intent to use it in their daily work, i.e. perform analysis, build models and so on. Maybe you are an analyst who wants to use more data science methods in your work.
2. Software developers implementing data science methods. Maybe you have a data science team that develops a model that you are responsible for implementing in production. It is important for everybody in the chain to understand why we do things a certain way.

## So, what is data science? 

Well, mostly it is about machine learning but absolutely not exclusively. Let us have a look at the data science process (when you are running solo):

## The Data Science Process
1. What question do you wish to answer? Does it make sense? Will it be beneficial? Are you allowed? (GDPR)
2. Locate the data, do you have it? Can you get it? Where are you allowed to process it?
3. Clean the data / transform the data. Most of the time the data is neither clean nor in the format that you would like.
4. Model the data, choose an appropriate algorithm. How can you validate your results?
5. Visualize your findings and illustrate key results.
6. Communicate your results to specialists and non-specialists.

### Is it this straightforward?
No, not really. Sometimes the steps above can be followed and the project is complete, but most times you will run into deviations from the above. A common deviation is that Step 1 does not originate from you but from some other part of the business. The problem this will pose for you as a data scientist is first of all to evaluate the question, i.e. does it makes sense from a data science perspective? The second problem it will pose for you is that, most often the stakeholder asking the question does not know what data is available and then it is your task to uncover it. This in turn can sometimes lead to the conclusion that the data to answer the original question is not available and therefore you need to circle back to the stakeholder to redefine the question. This process can take a substantial amount of time, not to mention the fact that in many bigger companies the data as well as the know-how to interpret it is scattered across multiple divisions.

#### Data cleaning

Data cleaning and transformation is typically a very big part of your work. Data cleaning is often a difficult and undefined process where you sometimes need to rely on either common sense or domain expertise to "fill in the blanks". 
Another common method is simply to exclude that data, which you can do if you still have a good amount of clean data. Data cleaning is actually part of the modeling process, and all choices that you make will impact the effectiveness of the models that you will develop later. It is paramount that you are involved in the [ETL](https://en.wikipedia.org/wiki/Extract,_transform,_load) process and actually have a hand in the data cleaning, even if this is delegated to someone else.

## Again, what is data science?

[Wikipedia](https://en.wikipedia.org/wiki/Data_science) puts it this way to paraphrase: "Data science is an interdisciplinary field that uses scientific methods, processes, algorithms and systems to extract knowledge and insights from data in various forms."

This is a very broad definition, we will focus on the particular interpretation

* Understanding the business
* Asking the right question
* Collecting the right data
* Building the "right" model
* Finding insights

### Asking the right question, i.e., is the answer useful?

It is very important that the answer that you seek will be useful to whatever the company is trying to accomplish. 

Some standard settings:

* Most often we will care about revenue or something which has a causal connection to the revenue.
* Solving a problem that will lead to a new service
* Solving a problem that will help the company satisfy legal requirements or just to improve the public opinion, like reducing environmental impact.

### Does the data help answer the question?

Sometimes when a project starts there is no information about what data is available, this will require some digging and eventually you may realize that you cannot give the answer needed with it. This is where the data scientist could really benefit from some knowledge about the problem at hand - for instance it could be business or it could be technical understanding. The goal when this happens is to make the best of the situation, i.e., is there something else that can be answered that will be useful with the available data?


Finding insights
Once we have asked a good question, obtained the appropriate data and estimated a model that is optimal at answering the question up to first order, what other insights can we draw from the model or even improve it? The following questions will help us take the next steps with the data science process:

* What is the impact of the prediction service provided by the estimated model based on changes to feature values?
    
    - “Features” here are nothing but what data or its transformations we have at our disposal to make a prediction about * an unobserved variable of interest using our estimated model.
* What are the most important features for the model? Does the model come with a way to rank its features by some notion of its importance in prediction?
* How confident are we about the estimated model and its predictions and what is the potential downstream impact if we use the model to make our decisions?



# CRISP-DM Model

The data mining process can be explained with the **CRISP-DM** model. CRISP-DM stands for _CRoss-Industry Standard Process for Data Mining_. As you probably will experience in your own Data Science project, the sequence of the phases is not rigid. We move back and forth between the different phases, the arrows in the picture indicate the most important and frequent movements.


<img src="CRISP_DM.jpg" width="500">

The outer circle symbolizes the cyclical nature of data mining itself, often the lessons learned during the process trigger new questions that benefits from the experiences of the previous ones.


## Business understanding

This initial phase focuses on understanding the project objectives and requirements from the business point of view. We then convert this knowledge into a data mining goal and a preliminary plan designed to achieve the objectives. What then is the difference between a business goal and a data mining goal?

A business goal states the objective in business terminology and a data mining goal states the project objective in technical terms. For example, the business goal might be “Increase market sales to existing customers.” A data mining goal might be “Predict how many widgets a customer will
buy, given their purchases over the past three years, demographic information (age, salary, city, etc.) and the price of the item.”

It is important to make sure that you understand the business question. In the worst case scenario you spend a great deal of effort producing the right answers to the wrong questions.

In this phase it can be good to describe the criteria for a successful, or useful outcome of the project from the business point of view. In the best case this is something specific that can be measured objectively, such as the reduction of customer churn to a certain level. Sometimes it is subjective such as “give useful insights into the relationships.” If you are given a business question such as the latter, a tip is to at least identify who makes this subjective judgment.

## Data understanding

The data understanding phase starts with the initial data collection. When we have the first data material we proceed with data exploration. In this step we familiarize ourselves with the data and identify data quality problems. This is done by querying, visualizing the data and report values. Example of steps: visualize variables, distribution of key attributes, relations between attributes, properties of significant sub-populations, simple statistical analyses. These analyses can directly address the data mining goals, but they can also contribute to the data description needed for further analysis. An important step in the data understanding is to examine the quality of the data. Try to figure out if the data contains errors (faulty sensor, etc.), and if there are errors how common they are. Are there any missing values in the data? 

## Data preparation

The data preparation phase covers all activities to construct the dataset that will be fed into the model. The work we do in this step will likely be performed multiple times, maybe we want to fit more models, or we discover something that leads us to redo the data preparation. The tasks in this phase include feature/attribute selection as well as transformation, cleaning and setting up the data to fit the model. Remember that all transformations you do to the data will impact the model and its results. In this step you can also consider if it makes sense to do feature engineering. In feature engineering you create new attributes, constructed from one or more existing attributes. Example: area = length * width.

## Modeling

In this phase, we try out various models (supervised learning, unsupervised learning, etc.), and calibrate the model parameters to their optimal values. Typically, there are several techniques for the same data mining problem, as the French poet Alain de Lille stated in the Middle Ages “all roads lead to Rome” which is true for Data Mining as well. Some models have specific requirements and make specific assumptions about the data, e.g., all attributes have uniform distributions, no missing values allowed, etc. Continue the model building and assessment of the parameters until you strongly believe that you have found the best model(s).

### Example 1 - looking for the number of unique entries

Let us now do some basic data exploration of our temperature data. We might for example want to know how many different weather stations are included in our dataset. 

We will look at two methods for answering this question. The first method relies mostly on basic Python functionality, and the other one makes use of the extended functionality provided by ``pandas``. You have seen some basic Python functionality in the Introduction to Python. Now we will demonstrate more of it, focusing on the usefulness of ``pandas``.

Let us now start with the first, basic Python method. To find out how many unique station names there are in the ``station_name`` column, we need to:

1. Walk through all entries of the column.
2. For each entry, we need to determine if we have seen this station name before.
    1. If this is a new name, we need to increase the counter of unique station names.
    2. If we already have seen this name, we can just move to the next entry of the column.

After we have iterated through the column, the value of the counter will be the number of unique station names in our dataset. Make sure that you see why this is true before you move on with the rest of this example.

````python
%%time
counter = 0
seen_stations = []
for station_name in temperatures_df['station_name']:
    if station_name not in seen_stations:
        counter = counter + 1
        seen_stations.append(station_name)
````

Now let us see how we implemented this method in Python. Look at the cell above. Please ignore ``%%time`` for now, this is the command that we use in order to show the execution time for later comparison of the two approaches.

The first thing that we need to do is to have a counter to keep track of the unique station names. Recall from the Introduction to Python, that we can use a variable in Python for this purpose, because the counter's value will change as the code is run:

```python
counter = 0
```

Remember that we are allowed to name variables freely (as long as we don't use Python keywords), but it is a good habit to use variable names that are descriptive. This makes the code much easier to understand. 

We could of course have used any value for the right-hand side of the assignment, but since we want ``counter`` to keep track of the number of unique station names we encounter, and we haven't started to count them yet, zero is the only value that makes sense.

When we are walking through the entries of the column, we need to be able to determine if we have encountered this station name before. The simplest way of doing this is to maintain a list of all station names we have seen. You have seen before how to create and manipulate a list object in Python, now let's use this knowledge. If you want an empty list (since you have not observed any station names yet), you just write the square brackets, as in the following:

```python
seen_stations = []
```

If we have a column entry with the variable name ``station_name``, we can test if it is in our list of seen station names using an ``if`` clause:

```python
if station_name not in seen_stations:
    counter = counter + 1
    seen_stations.append(station_name)
```

The first line of the above ``if`` block is pretty self-explanatory. If ``station_name`` is not in the list of seen stations, we have found a new station name, and the value of our counter needs to be increased. The second line of the ``if`` block adds (appends) ``station_name`` to the list of seen stations.

We now want to do this test for each entry in the ``station_name`` column. This can be done with a for loop:

```python
for station_name in temperatures_df['station_name']:
    if station_name not in seen_stations:
        counter = counter + 1
        seen_stations.append(station_name)
```
The name of the ``for`` variable is arbitrary, but as always, it is a good practice to use descriptive names.

When we combine all these parts we end up with the code in the cell above. 

To display the value of the counter, we can type the name of it in an empty cell.

````python
counter
````

Now when we have seen how to solve the station name counting problem using basic, "core", Python, let us see how we solve it using ``pandas``.

````python
%%time
temperatures_df['station_name'].nunique()
````

Note that not only is the ``pandas`` solution shorter to type, it is also faster to execute and easier to understand. We checked this using ``%%time`` on the first row of a cell.

For this reason it is a good idea to try to use as much built-in ``pandas`` functionality as possible when working with ``pandas`` objects. The reason why we looked at both methods is that if you know how to use the core Python functionality, you'll become a MacGyver of data analysis, and in the unlikely case that there is no ``pandas`` function that can solve your problem, ``pandas`` with some Python duct tape and rubber bands will most likely do the trick.

All Python objects have a collection of built-in functions you can use. We have already seen that ``pandas``  Series has a ``nunique`` function, but there are many more.

You might ask how you can learn them all by heart, and the answer is that you cannot and you should not. Instead, you need rely on your Googling skills and get good at reading the [documentation](https://pandas.pydata.org/pandas-docs/stable/).

### Filter, Sort and Groupby

Sometimes you might want to use certain condition(s) to filter the columns.

For example, ``temperatures_df[temperatures_df['date']>'2018-10-16']`` would give you only those rows of the dataframe where the column ``date`` contains the dates after 2018-10-16. You also can use ``&`` ("and") or ``|`` ("or") to add different conditions to the filtering and make it more advanced (it is called boolean filtering).

````python
temperatures_df[temperatures_df['date']>'2018-10-16']
````

````python
temperatures_df[(temperatures_df['date']>'2018-10-16')&(temperatures_df['date']<'2018-11-16')]
````

Sometimes it can be necessary to sort the values in an ascending or descending order based on certain column values. 
It is possible to do it an ascending order using ``df.sort_values(col1)``, and in a descending order using ``df.sort_values(col2,ascending=False)``. Moreover, one can sort values by one column in ascending order, then by another in descending order, using ``df.sort_values([col1,col2],ascending=[True,False])``.

Let's check how it works using our example. Recall that the dataframe looks like the following:

````python
temperatures_df
````

Let's now order it in ascending order by temperature.

````python
temperatures_df.sort_values('temperature')
````

Now in descending order by date.

````python
temperatures_df.sort_values('date',ascending=False)
````

Now let's order it ascending by date, but descending by time.

````python
temperatures_df.sort_values(['date','time'],ascending=[True,False])
````

One more useful command we consider is ``groupby``. This provides a functionality for splitting the data into groups based on some criteria, applying a function to each group independently and combining the results into a data structure. The command ``df.groupby(col)`` returns a groupby object for values from one column, however one can also use ``df.groupby([col1,col2])`` to get a result for several columns.

Let’s use ``groupby`` to calculate the mean temperature at each weather station.

````python
temperatures_df.groupby('station_name').mean()
````

### Example 2 - working with dates 

Let us now continue our analysis of the temperature dataset. In the previous example we found the number of unique stations in the data.

Another natural question to ask is: during which period was the data collected? We can obtain this information by asking for the minimum and maximum values of the column ``date``.

````python
temperatures_df['date'].min()
````

````python
temperatures_df['date'].max()
````

Recall that the columns ``date`` and ``time`` are strings, which are denoted as an ``object`` type in Python.

````python
temperatures_df.info()
````

 You can convert them both to the ``datetime `` type using ``pd.to_datetime() `` method. 
 
Why is it necessary to convert, you might ask? Because  ``datetime`` is a natural format for date and time in  ``pandas ``, providing some extra useful functionality when working with dates. For example, suppose we want to separate the month from the date in the very first row. If you try to do this with a string it is not going to work. 

![Trying to extract date from a string date](err6.png)

***

It does not work and you get an error message about a string not having the attribute month(). 

Let's now convert the ``date`` and ``time``  columns to the ``datetime`` format. We do the following steps:

1. We combine strings with dates and time to one string of the format "YYYY-MM-DD HH:MM:SS" (recall, that you can add strings together using "+")
2. Convert this string to the ``datetime`` format.

````python
temperatures_df['date'] = pd.to_datetime(temperatures_df['date']+' '+temperatures_df['time'])
````

Look at ``df.info()`` - now the ``date`` is not an ``object`` anymore.

````python
temperatures_df.info()
````

The column ``date`` has now changed.

````python
temperatures_df
````

It seems now that we don't need an extra column for time and can simply eliminate it using ``df.drop(name, axis)`` with ``axis=1``. It means that we drop the column. If you use ``axis=0``, it will drop a row with corresponding "name" in the index column.

````python
temperatures_df = temperatures_df.drop(['time'], axis=1)
temperatures_df
````

Let's now try to extract the month from the date on the first row.

````python
temperatures_df['date'][0].month
````

Now it works! 

Of course, ``month`` is not the only method one can use working with ``datetime`` format. For instance, one can return the day of the week. Note, that in ``pandas`` it is assumed, that the week starts on Monday, which is denoted by 0 and ends on Sunday which is denoted by 6.

````python
print(temperatures_df['date'][0])
print(temperatures_df['date'][0].dayofweek)
````
